```python
df['Has_Cabin'] = df['Cabin'].notnull().astype(int)  # 1 if cabin is known, 0 if NaN
print(df.groupby('Has_Cabin')['Survived'].mean())
```

```python
df['Deck'] = df['Cabin'].str[0]  # Extract first letter of Cabin
print(df.groupby('Deck')['Survived'].mean().sort_values())
```
